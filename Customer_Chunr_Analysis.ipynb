```python
# replacing the blacnk spaces with 0
df["TotalCharges"] = df["TotalCharges"].replace(" ", "0")

# cahnging the colum type from object to float
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

Info Gathered

1. As we have ealrier seen that the minimum tenure is 0 so we can deal with this by removing this, for this analysis this data is an outlier and not of any productive use.

2. Data in senior citizen is filled in boolean as 1 for true and 0 for false which creates a condusion and hence the count for senior citizen and the min and max is skewed.This needs to be fixed.

3. We have verified that this database does not have any duplicated records based on checking the database with the entire row and also with "customerID" which should be a unique column

```python
# creating a funciton to convert the senior citizen value from 0 adn 1 to Yes and No
def conv(value):
    if value == 1:
        return "Yes"
    else:
        return "No"

# now givng the dataframe of senior citizen to the funciton to change the values 
df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```

```python
#Validating the changes made above using sample to show random 20 records
df["SeniorCitizen"].sample(20)
```

We can see that out funciton has worked and we have yes and no instead of 0 and 1 in the data. This will make it easier for to visualise and get a perfect count of senior citizens


Based on this knowledge lets find out what percentage of subscribers are senior citizens

```python
#counting the total no of seniro citizens and total no. of not senior citizens
senior_count = df["SeniorCitizen"].value_counts()

#displaying the results
senior_count
```

```python
# Using the above information to visualise in pie chart
plt.figure(figsize = (4,4))
plt.pie(senior_count, labels = senior_count.index, autopct = '%1.1f%%', startangle = 90)
plt.title('Distribution of Senior Citizens')
plt.axis('equal') 
plt.show()
```

we can see that majority of the userbase is NOT seniro Citizen

Exploring the data to find out the percentage difference between the churn of senior citizen and non seniro citizen

```python
# Grouping the senior citizen and non seniro citizen and churned or not
senior_churn = df.groupby(['SeniorCitizen', 'Churn']).size().unstack()

# Convert counts to percentages row-wise
senior_churn_percent = senior_churn.div(senior_churn.sum(axis=1), axis=0) * 100

# Plot as stacked bar chart
ax = senior_churn_percent.plot(
    kind='bar',
    stacked=True,
    figsize=(4, 6),
)

#for adding the percentage in the bar 
for container in ax.containers:
    ax.bar_label(container, fmt='%1.1f%%', label_type='center')

#formatting the bar
plt.title('Churn Percentage: Senior vs Non-Senior Citizens')
plt.ylabel('Percentage')
plt.xlabel('Senior Citizen')
plt.xticks(rotation = 0)
plt.legend(title='Churn')
plt.show()
```

It can be clearly seen that even though the total % of senior citizens are just a quarter of the total population (refer to the distribution of senior citizen), within the senior citizen almost half of them has churned 

This is a point of concern as the percentage for non senior customer is relatively less

Now checking the distribution of customers based on the type of contract they have subscribed to

```python
plt.figure(figsize = (4,4))

ax = sns.countplot(
    x = 'Contract',
    data = df,
    hue = 'Churn'
)

total = len(df)

# Loop through each container (one for each hue level)
for container in ax.containers:
    ax.bar_label(
        container,
        labels=[f'{(v.get_height()/total*100):.1f}%' for v in container],  # % format
        label_type='edge'
    )

plt.title("Dstribution of Contracts")
plt.show()
```

This clearlys shows maximum of contracts are for Month-to-Month and also the maximum chruns have happened in the same contract

This has given us an understanding of what types of customers are churning out the most
